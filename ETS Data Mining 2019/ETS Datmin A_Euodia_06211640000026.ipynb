```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
online=pd.read_csv('online_retail.csv')
online.head()
```

```python
print(online.shape)
```

```python
np.sum(online.isnull())
```

```python
#Preprocessing Data
online.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
online['InvoiceNo'] = online['InvoiceNo'].astype('str') 
online=online[~online['InvoiceNo'].str.contains('C')]

online['UnitPrice'] = online['UnitPrice'].astype('str')
online=online[~online['UnitPrice'].str.contains('0')]
```

```python
np.sum(online.isnull())
```

```python
print(online.shape)
```

```python
a = online.groupby('InvoiceNo')
Transaksi = []
for i,j in a:
    Transaksi.append(list(j['Description'].map(str)))

transaksi=pd.DataFrame(Transaksi)
```

```python
transaksi
```

```python
transaksi.shape
```

```python
records=[]
for i in range (0, 19168):
    records.append([str(transaksi.values[i,j]) for j in range (0,923)])
```

```python
association_rules=apriori(records,min_support=0.01, 
                          min_confidence=0.2, 
                          min_lift=3, min_length=2) 
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
#firts index of the inner list
#contains base item and add item
    pair = item[0]
    items = [a for a in pair]
    print("rule: " + items [0] + " ->" + items[1])

    #second index of the inner list
    print("support: " + str(item[1]))

    #third index of the list located at 10th
    #of the third index of the inner list

    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=======================================")
```
