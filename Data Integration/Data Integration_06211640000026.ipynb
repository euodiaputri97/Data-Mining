##### Data tentang Heart Disease

```python
import numpy as np
import pandas as pd
```

```python
heart1=pd.read_csv('HEART1.csv')
heart2=pd.read_csv('HEART2.csv')
heart3=pd.read_csv('HEART3.csv')
heart4=pd.read_csv('HEART4.csv')

heart=pd.concat([heart1, heart2, heart3, heart4])
heart
```

```python
heart.to_csv( "heart.csv", index=False )
```

#### Data tentang Google Play Store

```python
playstore1=pd.read_csv('playstore1.csv')
playstore2=pd.read_csv('playstore2.csv')
playstore3=pd.read_csv('playstore3.csv')
playstore4=pd.read_csv('playstore4.csv')
playstore5=pd.read_csv('playstore5.csv')

playstore_1=pd.merge(playstore1, playstore2, on='App')
playstore_2=pd.merge(playstore3, playstore4, on='App')
playstore_3=pd.merge(playstore_1, playstore_2, on='App')
playstore=pd.merge(playstore_3, playstore5, on='App')
playstore
```

```python
playstore.to_csv( "playstore.csv", index=False )
```

####  Data tentang Luas Tanah (Hektar) berdasarkan Provinsi di Indonesia tahun 2003-2015

```python
hektarsawah1=pd.read_csv('hektarsawah1.csv')
hektarsawah2=pd.read_csv('hektarsawah2.csv')
hektarsawah3=pd.read_csv('hektarsawah3.csv')
hektarsawah4=pd.read_csv('hektarsawah4.csv')
hektarsawah5=pd.read_csv('hektarsawah5.csv')
hektarsawah6=pd.read_csv('hektarsawah6.csv')

hektarsawah_1=pd.merge(hektarsawah1, hektarsawah2,  on='Provinsi')
hektarsawah_2=pd.merge(hektarsawah3, hektarsawah4,  on='Provinsi')
hektarsawah_3=pd.merge(hektarsawah5, hektarsawah6,  on='Provinsi')
hektarsawah_4=pd.merge(hektarsawah_1, hektarsawah_2,  on='Provinsi')
hektarsawahfix=pd.merge(hektarsawah_4, hektarsawah_3,  on='Provinsi')
hektarsawahfix
```

```python
hektarsawahfix.to_csv( "hektarsawah.csv", index=False )
```
