### Mengatasi Missing Value

```python
import numpy as np
import pandas as pd
```

```python
real_es=pd.read_csv('Real estate.csv')
real_es.head()
```

```python
real_es.describe()
```

```python
np.sum(real_es.isnull())
```

```python
real_es['X1 house age']=real_es['X1 house age'].fillna((real_es['X1 house age'].mean()))
np.sum(real_es.isnull())
```

```python
real_es.to_csv( "real_es (tanpa missing value).csv", index=False ) #data tanpa missing value
```

### Mengatasi Data yang Missing Value dan Outlier

```python
from scipy import stats
```

```python
real_es1=pd.read_csv('real_es (tanpa missing value).csv') #data dengan missing value sudah diatasi
real_es1.head()
```

```python
z=np.abs(stats.zscore(real_es1._get_numeric_data()))
print(z)
```

```python
real_es1=real_es1[(z < 3).all(axis=1)]
real_es1.head()
```

```python
real_es1.to_csv( "real_es1 (tanpa missing value & outlier).csv", index=False ) #data tanpa missing value dan outlier
```

### Mengatasi Data yang Missing Value dan Melakukan Transformasi 

```python
import pandas as pd
import numpy as np
```

```python
real_es2=pd.read_csv('real_es (tanpa missing value).csv') #data dengan missing value sudah diatasi
real_es2=real_es2.iloc[0:415,1:7]
real_es2.head()
```

```python
#Normalisasi
from sklearn.preprocessing import normalize
real_es2_norm=normalize(real_es2,norm='l2') #default norm l2, lainnya L1 dan Max
real_es2_norm=pd.DataFrame(real_es2_norm)
real_es2_norm.head()
```

```python
real_es2_norm.to_csv( "real_es2 (tanpa missing value & ditransformasi).csv", index=False ) #data tanpa missing value dan ditransformasi
```
